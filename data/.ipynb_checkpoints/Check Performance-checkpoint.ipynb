```python
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

# Function to load Performance Dictionary

```python
def load_obj(name ):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
file_name = 'LSTM_POMDP_V4'
```

```python
perf = load_obj(name=file_name)
```

```python
def moving_average(data_set, periods=3):
    weights = np.ones(periods) / periods
    return np.convolve(data_set, weights, mode='valid')
```

```python
window_size = 500
```

# Loss Statistics


```python
#plt.plot(perf['loss'])
plt.plot(moving_average(perf['loss'],periods=window_size))
plt.xlabel('Steps')

plt.ylabel('Batch Loss')
plt.grid()
```

# Total Reward over episodes

```python
#plt.plot(perf['total_reward'])
plt.plot(moving_average(perf['total_reward'],periods=window_size),label='LSTM,BPTT=8')
plt.plot(mdp,label='MDP')
plt.plot(single,label='SINGLE FRAME')
plt.plot(two,label='TWO FRAMES')
plt.plot(four,label='FOUR FRAMES')
plt.xlabel('Episode')
plt.title('Epsilon Decreased over episodes till ,$\epsilon = 0.01$')
plt.ylabel('Episode Reward (Moving Avg, Window = %d)'%(window_size))
plt.legend()
plt.grid()
```

```python
mdp = moving_average(perf['total_reward'],periods=window_size)
```

```python
single = moving_average(perf['total_reward'],periods=window_size)
```
